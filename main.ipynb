```python
import sys
import os
sys.path.append(os.path.abspath(os.getcwd()))

try:
    %run src/data_loader.ipynb
    %run src/calculations.ipynb
    %run src/analysis.ipynb
    %run src/visualization.ipynb
    print("All modules loaded successfully.")
except Exception as e:
    print(f"Error loading modules: {str(e)}")
    sys.exit(1)
```

```python
def main():
    try:
        # Load and preprocess data
        sp500, euro50 = load_both_indices()
        print("Data loaded successfully.")
    except Exception as e:
        print(f"Error loading data: {str(e)}")
        return

    try:
        sp500 = calculate_returns(sp500)
        euro50 = calculate_returns(euro50)
        print("Returns calculated successfully.")
    except Exception as e:
        print(f"Error calculating returns: {str(e)}")
        return

    try:
        # Perform analysis
        sp500_analysis = analyze_index_performance(sp500)
        euro50_analysis = analyze_index_performance(euro50)
        comparison = compare_indices(sp500, euro50, 'S&P 500', 'EURO STOXX 50')
        correlation = correlation_analysis(sp500, euro50)
        print("Analysis completed successfully.")
    except Exception as e:
        print(f"Error during analysis: {str(e)}")
        return

    # Print results
    print("\nS&P 500 Analysis:")
    print(sp500_analysis)
    print("\nEURO STOXX 50 Analysis:")
    print(euro50_analysis)
    print("\nComparison:")
    print(comparison)
    print(f"\nCorrelation between S&P 500 and EURO STOXX 50: {correlation:.4f}")

    try:
        # Visualizations
        plot_returns_distribution(sp500, 'S&P 500')
        plot_returns_distribution(euro50, 'EURO STOXX 50')
        plot_cumulative_returns(sp500, euro50, 'S&P 500', 'EURO STOXX 50')
        plot_correlation_heatmap(sp500, euro50, 'S&P 500', 'EURO STOXX 50')
        plot_performance_comparison(comparison)
        print("Visualizations created successfully.")
    except Exception as e:
        print(f"Error creating visualizations: {str(e)}")
```

```python
if __name__ == "__main__":
    main()
```
